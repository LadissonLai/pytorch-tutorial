```python
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter("logs/")

for i in range(100):
    writer.add_scalar("Loss/train", i/2, i)  # 输入标题，因变量，自变量

writer.close()
```

```python
import cv2
writer = SummaryWriter("logs/")
img_path = "/home/txlinux/CL/pytorch-tutorial/imgs/003.jpg"
img_ndarray = cv2.imread(img_path)  # HWC
print(img_ndarray.shape)
writer.add_image("test img", img_ndarray, 1, dataformats='HWC')
writer.close()
```

总结：tensorboard使用方法
1. tensorboard用来记录训练过程，并进行可视化，包括loss，图片等等。
2. 使用方法：  
    1. 安装tensorboard：pip install tensorboard
    2. 运行命令：tensorboard --logdir=logs
    3. 打开浏览器，输入http://localhost:6006/即可看到可视化界面
    4. 注意：logdir是tensorboard记录的文件夹，需要先运行训练，然后运行tensorboard，最后在浏览器中打开。
    5. 注意：tensorboard记录的文件夹中，记录的文件是.tfevents文件，这个文件是二进制文件，不能直接打开。
3. writer.add_scalar()和writer.add_image()两个方法对应添加标量和图片。
